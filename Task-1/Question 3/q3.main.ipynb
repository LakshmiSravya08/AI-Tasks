```python
import pandas




def AtleastSixLength( data ) :
    
    six_length = []
    
    for x in data :

        if len( x ) >= 6 : six_length.append( x )
    
    return six_length
def RemovePunct( splitted ) :

   
    puncts = [ ',' ,'.', ':', ';', '"', "'" ] 

    for x in range( len(splitted) ) :

        while( splitted[ x ][-1] in puncts ) :

            splitted[x] = splitted[ x ][ : -1 ]

       

    
    for x in range( len(splitted) ) :

        while( splitted[ x ][0] in puncts ) :

            splitted[x] = splitted[ x ][ 1 : ]

       
    return splitted



def FindMostFreq( data ) :

    count = {}

    for i in data :

        if i not in count : 
            count[ i ] = 1

        else : 
            count[ i ] += 1    

    maxwords = []
    keys = count.keys()

    maxword = data[0]

    maxwords.append( maxword )
    print( maxwords )
    
    for i in keys :

        if ( ( count[i] == count[ maxwords[0] ] ) and i not in maxwords ) :

            maxwords.append( i )
        
        elif ( count[i] > count[ maxwords[0] ] ) :
            maxwords = [ ]
            maxwords.append( i )
                
    return maxwords



about = ""
with open( "about.txt" ,'r' ) as textfile :
    about = textfile.read()

splitted = about.split()
rem_punct = RemovePunct( splitted )


print( "Words with atleast 6 letters are as follows : ", pandas.unique( AtleastSixLength( rem_punct ) ) )

print( "Most occuring word is : ", FindMostFreq( rem_punct ) )
```
