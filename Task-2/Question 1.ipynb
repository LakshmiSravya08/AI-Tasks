```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv( "Axisbank.csv" )
df
```

```python
df = df.truncate(after = 30)
df
```

```python
b_plot = sns.barplot( x = df["Date"], y = df["Volume"] )
plt.draw()
b_plot.set_xticklabels( b_plot.get_xticklabels() , rotation = 90 )
print()
```

```python
h_plot = sns.histplot( x = df["Date"], y = df["High"] )
plt.draw()
h_plot.set_xticklabels( h_plot.get_xticklabels() , rotation = 90 )
print()
```

```python
sc_plot = sns.scatterplot( x = df["Date"], y = df["Low"] )
plt.draw()
sc_plot.set_xticklabels( sc_plot.get_xticklabels() , rotation = 90 )
print()
```
