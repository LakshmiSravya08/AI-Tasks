```python
import seaborn as sns
import pandas as pd
```

```python
df = pd.read_csv( "country_wise_latest.csv" )
df = df.truncate( after=50 )
df
```

```python
covid_plot = sns.histplot( x = df["Country/Region"] , y = df["Confirmed"] )
covid_plot.set_xticklabels( covid_plot.get_xticklabels( ) , rotation = 90 )
print()
```
